```python
# subplots in seaborn
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df = sb.load_dataset('tips')
print(df.to_string())
g = sb.FacetGrid(df, col = "time")
g.map(plt.hist, "day")
plt.show()
```

```python
g = sb.FacetGrid(df, col = "sex", hue = "day")
g.map(plt.scatter, "total_bill", "tip")
plt.show()
```

```python
# statistical estimation
sb.barplot(x = "sex", y = "tip", hue = "time", data = df)
plt.show()
```

```python
#linear relationship
sb.regplot(x = "total_bill", y = "tip", data = df)
plt.show()
```

```python
#pair plot
sb.pairplot(df)
plt.show()
```

```python
# plotting selected variables
selected_vars = ['total_bill', 'tip']
sb.pairplot(df, vars=selected_vars)
plt.show()
```

```python
# adding a hue color to a seaborn pairplot
sb.pairplot(df,hue = 'size')
plt.show()
```

```python
#modifying color palette
sb.pairplot(df, hue="size",palette="husl")
plt.show()
```
